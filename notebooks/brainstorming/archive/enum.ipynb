```python
def naturals_gen():
    n = 0
    while True:
        yield n
        n += 1
```

```python
def integers_gen():
    naturals = naturals_gen()
    n = next(naturals)
    i = n
    while True:
        yield i
        n = next(naturals)
        i = i + (n * ((-1) ** (n - 1)))
```

I want to generate every list of integers.

```python
def integer_sequences_gen(n: int, length: int) -> list[list[int]]:
    if length == 0:
        return [[n]]

    seqs: list[list[int]] = []

    for i in range(5):
        seqs += integer_sequences_gen(i, length - 1)

    for seq in seqs:
        seq.insert(0, n)

    return seqs


print(integer_sequences_gen(None, 2))
```
