## Test 1: Basic usage

```python
from harmonica.time._onsetfunc import OnsetFunc
from harmonica.utility._gm import GMDrum
from harmonica.utility._mixed import Mixed
from harmonica.velocity import VelocityFunc


rhythm_pat = OnsetFunc([Mixed("1/4")])
rhythm = rhythm_pat.to_clip(duration=Mixed(64), drum=GMDrum.HiWoodBlock)
vfunc = VelocityFunc(pattern=[0, 1, 0, 2, 0, 0, 1], resolution=Mixed("1/4"))

velocities = [Mixed(x) for x in ["1/3", "2/3", 1]]

for event in rhythm.get_drum_events():
    event.velocity = vfunc.evaluate(event.onset, velocities)

# rhythm.preview()
```

So far so good.
