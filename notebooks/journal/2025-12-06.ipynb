# Time functions

Time functions are cool. They represent rhythmic patterns.

Let's look at some kicks.

```python
from harmonica.time._timefunc import OnsetFunc
from harmonica.utility._gm import GMDrum
from harmonica.utility._mixed import Mixed


kick_pattern = OnsetFunc([Mixed(x) for x in ["3/4", "1 1/2", 2]]).to_clip(
    Mixed(32), drum=GMDrum.BassDrum1
)

# kick_pattern.preview(tempo=140)
```

Let's add some snares.

```python
from harmonica.time._event._clip import Clip


snare_pattern = OnsetFunc([Mixed(4)], offset=Mixed(2)).to_clip(
    Mixed(32), drum=GMDrum.ElectricSnare
)

# Clip([kick_pattern, snare_pattern]).preview(tempo=140)
```

Let's add some hats, with a stress pattern applied to the velocities.

```python
from itertools import cycle


hats_pattern = OnsetFunc([Mixed("1/4")]).to_clip(Mixed(32), drum=GMDrum.ClosedHiHat)
stress_pattern = cycle([Mixed(x) for x in [1, "1/2", "1/4", 1, "1/3"]])
for hat in hats_pattern.get_drum_events():
    hat.velocity = next(stress_pattern)

Clip([kick_pattern, snare_pattern, hats_pattern]).preview(tempo=140)
```
