# Previewing rhythm parts

I need a way of previewing rhythm parts. I already have an enum class set up for specifying drum sounds in the GM standard.

I am working specifically with the Microsoft GS Wavetable Synth, so I need to set up drum tracks for that specifically.

Here is the strategy I will use: I will reserve channel 10 for drums.

There will be a kind of clip called a DrumClip. 

DrumClips contain DrumEvents, and has its own preview function.

A DrumEvent consists of an onset, a drum_piece and a velocity. drum_piece is an int and maps to a drum piece in the GM standard drum kit.

## Attempting to construct and preview a drum track

This will be tedious, as I will have to do it by hand. Here goes...

```python
from harmonica.time._event._clip import Clip, DrumClip
from harmonica.time._event._event import DrumEvent
from harmonica.utility import GMDrum, Mixed


kick_onsets = [Mixed(x) for x in [0, 1, 2, 3, 4, 5, 6, 7]]
kick_clip = DrumClip([DrumEvent(onset, GMDrum.BassDrum1) for onset in kick_onsets])

clave_onsets = [Mixed("4/7") * x for x in range(21)]
clave_clip = DrumClip([DrumEvent(onset, GMDrum.Claves) for onset in clave_onsets])

Clip([kick_clip, clave_clip]).preview()
```
